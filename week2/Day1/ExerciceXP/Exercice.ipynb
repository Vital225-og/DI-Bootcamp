# Exercice 1 : Introduction à l’analyse des données

## Introduction

Aujourd’hui, les données occupent une place très importante dans presque tous les domaines : les entreprises, les banques, les hôpitaux, les écoles, les gouvernements et même les plateformes numériques produisent chaque jour une grande quantité d’informations. Cependant, ces données ne sont réellement utiles que lorsqu’elles sont bien collectées, organisées, analysées et interprétées. C’est dans ce contexte que l’analyse de données devient essentielle.

## 1. Qu’est-ce que l’analyse de données ?

L’analyse de données est un processus qui consiste à collecter, nettoyer, organiser, étudier et interpréter des données afin d’en tirer des informations utiles. Elle permet de transformer des données brutes en connaissances exploitables pour mieux comprendre une situation, identifier des tendances, résoudre des problèmes ou prendre de meilleures décisions.

Par exemple, une entreprise peut analyser les données de ses ventes pour savoir quels produits sont les plus achetés, quelles périodes sont les plus rentables ou quels clients sont les plus actifs. Dans le domaine financier, l’analyse de données peut aussi aider à détecter des transactions suspectes ou à prévenir les fraudes.

## 2. Pourquoi l’analyse des données est-elle importante dans les contextes modernes ?

L’analyse des données est importante parce que le monde actuel fonctionne de plus en plus avec l’information. Les organisations prennent de nombreuses décisions à partir des données : choix commerciaux, amélioration des services, gestion des risques, personnalisation des offres, prévision des besoins et optimisation des ressources.

Elle permet notamment de :

- prendre des décisions plus rapides et plus fiables ;
- comprendre le comportement des clients ou des utilisateurs ;
- détecter des erreurs, des risques ou des fraudes ;
- améliorer les performances d’une organisation ;
- anticiper les tendances futures grâce aux données historiques.

Aujourd’hui, les tendances modernes montrent que l’analyse de données est fortement liée à l’intelligence artificielle, au machine learning, à l’analyse en temps réel et à la visualisation des données. Les entreprises utilisent de plus en plus les données pour prévoir la demande, personnaliser les services et automatiser certaines décisions.

## 3. Trois domaines d’application actuels de l’analyse de données

### 3.1 Finance et banque

Dans le domaine financier, l’analyse de données est utilisée pour surveiller les transactions, détecter les fraudes, évaluer les risques et mieux comprendre les habitudes des clients. Par exemple, une banque ou un service Mobile Money peut analyser les montants, les horaires, les lieux et la fréquence des transactions pour repérer des comportements suspects.

Dans un projet comme une plateforme de détection de fraude Mobile Money, l’analyse de données permettrait d’identifier les opérations anormales, de calculer des scores de risque et d’aider les analystes à prendre de meilleures décisions.

### 3.2 Santé

Dans le domaine de la santé, l’analyse de données aide les hôpitaux et les professionnels médicaux à mieux suivre les patients, prévoir certaines maladies, améliorer les traitements et optimiser la gestion des ressources. Les données peuvent provenir des dossiers médicaux, des résultats d’examens, des consultations ou des systèmes hospitaliers.

Par exemple, l’analyse de données peut aider à identifier les patients à risque, à suivre l’évolution d’une maladie ou à améliorer l’organisation des services médicaux.

### 3.3 Commerce et marketing

Dans le commerce et le marketing, l’analyse de données permet de comprendre les préférences des clients, de recommander des produits, de prévoir les ventes et d’améliorer les campagnes publicitaires. Les entreprises analysent les achats, les visites sur les sites web, les avis clients et les comportements sur les réseaux sociaux.

Par exemple, une boutique en ligne peut utiliser l’analyse de données pour savoir quels produits intéressent le plus les clients, quels articles sont souvent achetés ensemble et à quel moment lancer une promotion.

## Conclusion

En conclusion, l’analyse de données est devenue indispensable dans les contextes modernes. Elle permet de transformer des données brutes en informations utiles pour prendre de meilleures décisions. Que ce soit dans la finance, la santé, le commerce ou d’autres domaines, elle aide les organisations à comprendre leur environnement, à améliorer leurs services et à anticiper les évolutions futures. Avec le développement de l’intelligence artificielle et du machine learning, l’analyse de données devient une compétence essentielle pour les professionnels du numérique.

# Exercice 3 : Identification des types de données

Dans cet exercice, chaque colonne des datasets précédemment chargés est classée comme quantitative ou qualitative.

Une donnée quantitative est une donnée numérique qui peut être mesurée ou utilisée dans des calculs statistiques.

Une donnée qualitative est une donnée qui décrit une catégorie, un groupe, un statut ou une information textuelle.

```python
import pandas as pd

def classifier_colonnes(df, nom_dataset):
    resultats = []

    for colonne in df.columns:
        type_pandas = df[colonne].dtype
        valeurs_uniques = df[colonne].nunique()

        if pd.api.types.is_numeric_dtype(df[colonne]):
            classification = "Quantitative"
            raison = "La colonne contient des valeurs numériques qui peuvent être mesurées ou utilisées dans des calculs."
        else:
            classification = "Qualitative"
            raison = "La colonne contient des catégories, du texte ou des informations descriptives."

        resultats.append({
            "Dataset": nom_dataset,
            "Colonne": colonne,
            "Type Pandas": type_pandas,
            "Classification": classification,
            "Raison": raison
        })

    return pd.DataFrame(resultats)




print("Colonnes du dataset Sommeil :")
print(sleep_df.columns)

print("\nColonnes du dataset Santé mentale :")
print(mental_health_df.columns)

print("\nColonnes du dataset Carte de crédit :")
print(credit_card_df.columns)



classification_sleep = classifier_colonnes(
    sleep_df,
    "How Much Sleep Do Americans Really Get?"
)

classification_mental = classifier_colonnes(
    mental_health_df,
    "Global Trends in Mental Health Disorder"
)

classification_credit = classifier_colonnes(
    credit_card_df,
    "Credit Card Approvals"
)

classification_complete = pd.concat(
    [classification_sleep, classification_mental, classification_credit],
    ignore_index=True
)

classification_complete
```

# Exercice 9 : Exporter un DataFrame au format Excel et JSON

Dans cet exercice, nous créons un DataFrame simple avec Pandas, puis nous l’exportons sous deux formats différents : Excel et JSON. Cette opération permet de sauvegarder des données analysées ou préparées afin de les réutiliser dans d’autres outils ou applications.

```python
!pip install numpy
!pip install faker
```

```python
import numpy as np
import pandas as pd
from faker import Faker

fake = Faker()

# Génération de noms, âges et sexes fictifs
data = [
    (
        fake.name(),
        np.random.randint(18, 70),
        np.random.choice(["Male", "Female"])
    )
    for _ in range(100)
]

# Création du DataFrame
df = pd.DataFrame(data, columns=["Name", "Age", "Sex"])

df.head()
df.to_excel("personnes_fictives.xlsx", index=False)

df.to_json("personnes_fictives.json", orient="records", indent=4)
```

## Conclusion

Dans cet exercice, un DataFrame simple a été créé avec Pandas à partir de données fictives générées avec Faker et NumPy. Le DataFrame contient des noms, des âges et le sexe des personnes. Il a ensuite été exporté au format Excel avec `to_excel()` et au format JSON avec `to_json()`. Ces formats permettent de sauvegarder et de partager facilement les données.
