## Gestion et analyse des données en Python

Dans ce défi quotidien, nous travaillons avec un dataset sur les salaires des emplois en science des données.

## Normalisation Min-Max

La normalisation Min-Max permet de transformer les valeurs d’une colonne numérique dans une plage comprise entre 0 et 1.  
Dans cet exercice, la colonne `salary` a été normalisée afin de rendre les valeurs salariales comparables sur une même échelle.

La nouvelle colonne `salary_normalized` contient donc les salaires transformés entre 0 et 1.

```python
minmax_scaler = MinMaxScaler()

df["salary_normalized"] = minmax_scaler.fit_transform(df[["salary"]])

df[["salary", "salary_normalized"]].head()

```

## Réduction de dimensionnalité avec PCA

L’Analyse en Composantes Principales, ou PCA, permet de réduire le nombre de variables numériques tout en conservant une partie importante de l’information contenue dans le dataset.

Dans cet exercice, les colonnes numériques ont d’abord été standardisées, car PCA est sensible aux différences d’échelle entre les variables. Ensuite, PCA a été appliqué pour réduire les données à deux composantes principales : `PC1` et `PC2`.

Ces deux composantes permettent de représenter les données dans un espace plus simple, tout en conservant une partie de la variance du dataset.

```python
numeric_df = df.select_dtypes(include=["int64", "float64"])

numeric_df.head()
```

```python
standard_scaler = StandardScaler()

numeric_scaled = standard_scaler.fit_transform(numeric_df)

numeric_scaled = pd.DataFrame(numeric_scaled, columns=numeric_df.columns)

numeric_scaled.head()
```

```python
pca = PCA(n_components=2)

pca_result = pca.fit_transform(numeric_scaled)

pca_df = pd.DataFrame(pca_result, columns=["PC1", "PC2"])

pca_df.head()
```

```python
print("Variance expliquée par PC1 et PC2 :", pca.explained_variance_ratio_)
print("Variance totale conservée :", pca.explained_variance_ratio_.sum())
```

```python
salary_by_experience = df.groupby("experience_level")["salary"].agg(
    salaire_moyen="mean",
    salaire_median="median"
).round(2)

salary_by_experience
```
