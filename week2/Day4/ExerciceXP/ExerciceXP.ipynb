Exercice 1 expliquons l'importance de l'analyse des données

## Pourquoi la visualisation des données est-elle importante dans l’analyse des données ?

La visualisation des données est importante parce qu’elle permet de transformer des données brutes, souvent complexes, en représentations visuelles faciles à comprendre. Grâce aux graphiques, aux courbes, aux diagrammes et aux tableaux de bord, il devient plus simple d’identifier des tendances, de comparer des informations et de prendre des décisions.

### 1. Faciliter la compréhension des données

Les données sous forme de tableaux peuvent parfois être difficiles à interpréter, surtout lorsqu’elles sont nombreuses. La visualisation permet de présenter ces informations de manière plus claire et plus rapide. Par exemple, un graphique peut montrer en quelques secondes l’évolution des ventes, des accidents, des salaires ou des performances d’un projet.

### 2. Identifier les tendances et les relations

Les visualisations permettent de repérer facilement les tendances dans le temps, les évolutions, les corrélations entre variables ou les différences entre plusieurs groupes. Par exemple, une courbe peut montrer si un phénomène augmente ou diminue au fil des années, tandis qu’un diagramme de dispersion peut aider à observer une relation entre deux variables.

### 3. Détecter les anomalies

Les graphiques permettent aussi de repérer les valeurs inhabituelles, appelées valeurs aberrantes ou outliers. Ces anomalies peuvent révéler des erreurs dans les données, des comportements exceptionnels ou des situations importantes à analyser plus profondément.

### 4. Aider à la prise de décision

La visualisation des données aide les décideurs à comprendre rapidement les informations importantes. Les tableaux de bord permettent de regrouper plusieurs indicateurs clés au même endroit. Cela facilite la prise de décision basée sur des faits et non sur des impressions.

### 5. Communiquer les résultats

Enfin, la visualisation est un outil puissant de communication. Elle permet de présenter les résultats d’une analyse à un public technique ou non technique. Grâce au data storytelling, les données peuvent être utilisées pour raconter une histoire claire, convaincre un public et soutenir une recommandation.

## Conclusion

En résumé, la visualisation des données est essentielle dans l’analyse de données, car elle facilite la compréhension, révèle les tendances, met en évidence les anomalies et aide à prendre de meilleures décisions. Elle transforme les données en informations visuelles utiles et exploitables.

## Décrire l'utilité d'un graphique linéaire en visualisation de données.
Un graphique linéaire est utile en visualisation de données parce qu’il permet de montrer l’évolution d’une variable sur une période donnée ou selon un ordre précis.

Il est surtout utilisé pour représenter des données dans le temps, comme l’évolution des ventes par mois, du nombre d’accidents par année, de la température au fil des jours ou du chiffre d’affaires d’une entreprise.

Grâce à la ligne qui relie les points, on peut facilement observer les tendances générales : augmentation, diminution, stabilité ou variation brusque. Le graphique linéaire permet aussi de repérer les pics, les baisses importantes et les changements de comportement dans les données.

En résumé, un graphique linéaire est particulièrement adapté pour suivre une évolution et comprendre rapidement comment une donnée change dans le temps.

Exercice 2 : Création d’un graphique linéaire pour la variation de température

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

y = np.linspace(-1, 30, 7)
x = ["Lundi", "Mardi", "Mercredi", "Jeudi", "Vendredi", "Samedi", "Dimanche"]
plt.xlabel("Jour de la semaine")
plt.ylabel("Température (°C)")
plt.title("Température par jour de la semaine")
plt.plot(x, y, marker="o")
plt.show()

```

## Exercise 4: Visualizing the Distribution of CGPA

##  Exercice 5 : Comparaison des niveaux d’anxiété selon le sexe

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

student_df = pd.read_csv("Student Mental health.csv")

anxiety_df = student_df[["Choose your gender", "Do you have Anxiety?"]].copy()

anxiety_df["Anxiety_Numeric"] = anxiety_df["Do you have Anxiety?"].map({
    "Yes": 1,
    "No": 0
})

sns.barplot(
    data=anxiety_df,
    x="Choose your gender",
    y="Anxiety_Numeric",
    palette="viridis"
)

plt.title("Comparaison du niveau d'anxiété selon le sexe")
plt.xlabel("Genre")
plt.ylabel("Proportion d'étudiants ayant de l'anxiété")
plt.show()
```
