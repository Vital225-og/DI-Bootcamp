Exercice 3 : Comprendre les distributions

Objectif : générer une distribution normale avec moyenne 50 et écart type 10.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mean = 50
std = 10

x = np.linspace(mean - 4*std, mean + 4*std, 100)
y = norm.pdf(x, mean, std)

plt.plot(x, y)
plt.title("Distribution normale")
plt.xlabel("Valeurs")
plt.ylabel("Densité de probabilité")
plt.show()
```

Une distribution normale est une distribution en forme de cloche. La majorité des valeurs se trouvent autour de la moyenne. Ici, la moyenne est 50 et l’écart type est 10, ce qui signifie que les valeurs sont généralement réparties autour de 50.

Exercice 4 : Application du test t

```python
import numpy as np
from scipy import stats

np.random.seed(42)

data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)

t_stat, p_value = stats.ttest_ind(data1, data2)

print("Statistique t :", t_stat)
print("Valeur p :", p_value)
```

```python
if p_value < 0.05:
    print("Les deux groupes sont significativement différents.")
else:
    print("Les deux groupes ne sont pas significativement différents.")
```

Exercice 5 : Régression linéaire

```python
import numpy as np
import scipy.stats as stats

house_sizes = np.array( [50, 70, 80, 100, 120])
house_prices = np.array([150000, 200000, 210000, 250000, 280000])

slope, intercept, r_value, p_value, std_err = stats.linregress(house_sizes, house_prices)

print("Pente :", slope)
print("Ordonnée à l'origine :", intercept)
print("Coefficient de corrélation r :", r_value)
print("Valeur p :", p_value)
print("Erreur standard :", std_err)
```

predire le prix d'une maison de 90 m²

```python
house_size = 90
predicted_price = slope * house_size + intercept

print("Prix estimé pour une maison de 90 m² :", predicted_price)
```

```python
plt.scatter(house_sizes, house_prices, label="Données réelles")

predicted_prices = slope * house_sizes + intercept
plt.plot(house_sizes, predicted_prices, label="Droite de régression")

plt.title("Régression linéaire : taille vs prix")
plt.xlabel("Taille de la maison")
plt.ylabel("Prix")
plt.legend()
plt.show()
```

La pente représente l’augmentation estimée du prix lorsque la taille de la maison augmente d’un mètre carré. Si la pente est positive, cela signifie que plus la maison est grande, plus son prix estimé est élevé.

Exercice 6 : ANOVA

```python
from scipy import stats

fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]

f_stat, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)

print("Valeur F :", f_stat)
print("Valeur p :", p_value)
```

```python
if p_value < 0.05:
    print("Il existe une différence significative entre les effets des engrais.")
else:
    print("Il n'existe pas de différence significative entre les effets des engrais.")
```

L’ANOVA permet de comparer les moyennes de plusieurs groupes. Si la valeur p est inférieure à 0,05, cela signifie qu’au moins un groupe est significativement différent des autres. Si la valeur p est supérieure à 0,05, on ne peut pas conclure qu’il existe une différence significative entre les groupes.

Exercice 7 : Distribution binomiale

```python
from scipy.stats import binom

n = 10
p = 0.5
k = 5

probability = binom.pmf(k, n, p)

print("Probabilité d'obtenir exactement 5 faces en 10 lancers :", probability)
```

La distribution binomiale permet de calculer la probabilité d’obtenir un certain nombre de succès dans un nombre fixe d’essais. Ici, un succès correspond à obtenir face lors d’un lancer de pièce.

Exercice 8 : Corrélation Pearson et Spearman

```python
import pandas as pd
from scipy import stats

data = pd.DataFrame({
    'age': [23, 25, 30, 35, 40],
    'income': [35000, 40000, 50000, 60000, 70000]
})

pearson_corr, pearson_p = stats.pearsonr(data["age"], data["income"])
spearman_corr, spearman_p = stats.spearmanr(data["age"], data["income"])

print("Corrélation de Pearson :", pearson_corr)
print("p-value Pearson :", pearson_p)

print("Corrélation de Spearman :", spearman_corr)
print("p-value Spearman :", spearman_p)
```

Le coefficient de corrélation mesure la force de la relation entre deux variables. Une valeur proche de 1 indique une forte relation positive, une valeur proche de -1 indique une forte relation négative, et une valeur proche de 0 indique une faible relation.
